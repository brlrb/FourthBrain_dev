```python
import pandas as pd
```

```python
tsla_df_raw = pd.read_csv('TSLASentimentAnalyzer/sentiment_data.csv')
ford_df_raw = pd.read_csv('TSLASentimentAnalyzer/Fordstock_sentiment_data.csv')
```

```python
tsla_df['stock_ticker'] = "TSLA"
ford_df['stock_ticker'] = "FORD"
```
